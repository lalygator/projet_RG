On veut simuler la trajectoire d'une étoile proche d'un trou noir de Schwarzschild. 





Pour ça il faut calculer avec un pas de temps :

- la distance de l'étoile du trou noir à un instant
- la vitesse angulaire $V_\varphi$
- la vitesse rayon $V_r$
- l'angle de déplacement $\varphi$

### Facteur $\frac{dt}{ds}=\frac{dt}{d\tau}=\dot{t_0}$

Soit la métrique de Schwarzschild dans un plan défini par $\theta = \frac{1}{2}$, on a :

\begin{equation}

ds^2 = -\left(1-\frac{2GM}{c^2 r}\right)dt^2+\left(1-\frac{2GM}{c^2r}\right)^{-1}dr^2+r^2d\varphi^2

\end{equation}

Pour trouver l'expression $\frac{dt}{ds}$, on réexprime la métrique tel que :

\begin{align*}

ds^2 &= -\left(1-\frac{2GM}{c^2 r}\right)dt^2+\left(1-\frac{2GM}{c^2r}\right)^{-1}dr^2+r^2d\varphi^2 \\

\iff \frac{ds^2}{dt^2} &= -\left(1-\frac{2GM}{c^2 r}\right)+\left(1-\frac{2GM}{c^2r}\right)^{-1}\frac{dr^2}{dt^2}+r^2\frac{d\varphi^2}{dt^2}

\end{align*}

Or on a $\left(\frac{dr}{dt}\right)^2=V_r^2$ et $\left(\frac{d\varphi}{dt}\right)^2=V_\varphi^2$, d'où :

\begin{align*}

\frac{ds^2}{dt^2} &= -\left(1-\frac{2GM}{c^2 r}\right)+\left(1-\frac{2GM}{c^2r}\right)^{-1}V_r^2+r^2 V_\varphi^2 \\ \: \\

\iff \frac{dt}{ds} &= \frac{1}{\sqrt{-\left(1-\frac{2GM}{c^2 r}\right)+\left(1-\frac{2GM}{c^2r}\right)^{-1}V_r^2+r^2 V_\varphi^2}} \\ 
    &= \frac{1}{\sqrt{\left(1-\frac{2GM}{c^2 r}\right)-\left(1-\frac{2GM}{c^2r}\right)^{-1}V_r^2-r^2 V_\varphi^2}}
\end{align*}

## Orbite stable

On définit une orbite stable toute orbite telle que

Pour recherche les paramètres d'orbite stable de notre trou noir, on impose un moment cinétique $L$ et la masse de trou noir $M$, et on résoud pour les racines de l'équation 

\begin{equation}\frac{dU}{dr}=\frac{M}{r^2}-\frac{L^2}{M^3}+\frac{3ML^2}{r^4}=0\end{equation}

que l'on peut réexprimer comme

\begin{equation}Mr^2-L^2M^3r+3ML^2=0\end{equation}

En imposant $L^2>12M^2$, on trouve comme solution

\begin{cases}

r_+ = \frac{L^2+\sqrt{L^4 - 12M^2L^2}}{2M}
\\
r_- = \frac{L^2-\sqrt{L^4 - 12M^2L^2}}{2M}

\end{cases}

Si l'on se trouve légèrement en dessous de $r_-$, alors l'objet tombe dans le trou noir et si on se trouve légèremnet au-dessus de $r_+$, alors l'orbite va osciller. Ceci dit, notre modèle considère que le corps sera toujours nécessairment attiré par le trou noir


## Mesure des énergies cinétique et potentielle

On calcule l'énergie potentielle 

$$E_p= \left(\frac{2M}{r} - 1\right) \times \left(-\frac{1}{2}- \mathcal{L}\left(\frac{L}{r}\right)^2\right)$$

## Calcul du redshift

Redshift gravitationnel

In the theory of general relativity, there is time dilation within a gravitational well. This is known as the gravitational redshift or Einstein Shift.[45] The theoretical derivation of this effect follows from the Schwarzschild solution of the Einstein equations which yields the following formula for redshift associated with a photon traveling in the gravitational field of an uncharged, nonrotating, spherically symmetric mass


$$ 1 + z = \frac{1}{1-\sqrt{\dfrac{2GM}{rc^2}}}  = \frac{1}{1-\sqrt{\dfrac{R_S}{r}}} $$
avec $R_S= \dfrac{2GM}{c^2}$ le rayon de Schwarzschild

Comme on considère ici notre $M$ un facteur de $M_0=2\cdot 10^{30}$ la masse du soleil, et $G=c=1$, on calcule dans notre code le redshift tel que

$$ z = \frac{1}{1-\sqrt{\dfrac{2M}{r}}} - 1 $$
